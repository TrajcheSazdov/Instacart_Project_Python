```python
# Follow the instructions in the Exercise for creating a loyalty flag for existing 
# customers using the transform() and loc() functions.
```

```python
# Grouping the dataframe by user_id and order_number and creating new column max_order

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# Adding loyalty_flag Loyal customer

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Adding loyalty_flag Regular customer

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Adding loyalty_flag New customer

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking if right flags ware assigned

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(15)
```

```python
# 05. The marketing team at Instacart wants to know whether there’s a difference between the 
# spending habits of the three types of customers you identified. 
# Use the loyalty flag you created and check the basic statistics of the product prices for each 
# loyalty category (Loyal Customer, Regular Customer, and New Customer). 
# What you’re trying to determine is whether the prices of products purchased by loyal customers differ 
# from those purchased by regular or new customers.
```

```python
# Checking the basic statistics of the product prices for each loyalty category

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# The average price of products purchased by Loyal Customers is lower than the products purchased by Regular or New Customers

```python
# 06. The team now wants to target different types of spenders in their marketing campaigns. 
# This can be achieved by looking at the prices of the items people are buying. 
# Create a spending flag for each user based on the average price across all their orders using the following criteria:

# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

```python
# Grouping the dataframe by user_id and prices and creating new column average_price

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# Adding spenders_flag Low spender

ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spenders_flag'] = 'Low spender'
```

```python
# Adding spenders_flag High spender

ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spenders_flag'] = 'High spender'
```

```python
ords_prods_merge['spenders_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head()
```

```python
# 07.In order to send relevant notifications to users within the app 
# (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent 
# versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior 
# according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
# Grouping the dataframe by user_id and days_since_prior_order and creating new column frequency_median

ords_prods_merge['frequency_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Adding frequency_flag Non-frequent customer

ords_prods_merge.loc[ords_prods_merge['frequency_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Adding frequency_flag Regular customer

ords_prods_merge.loc[(ords_prods_merge['frequency_median'] > 10) & (ords_prods_merge['frequency_median'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Adding frequency_flag Frequent customer

ords_prods_merge.loc[ords_prods_merge['frequency_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
# Marking 5,127 rows with outliers as missing since  they don’t make sense in terms of the other values in the column

ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
